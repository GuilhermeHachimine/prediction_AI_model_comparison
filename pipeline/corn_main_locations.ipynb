```python
import pandas as pd
import os


# Ler o arquivo CSV usando o caminho relativo
df = pd.read_csv('../data/corn_prices.csv')

grouped = df.groupby('main_location_id')
grouped.head()
```

```python
for main_location, group in grouped:
    # Criar o diretório se não existir
    dir_path = f'../data/corn/{main_location}'
    os.makedirs(dir_path, exist_ok=True)
    
    # Salvar o arquivo CSV
    file_path = os.path.join(dir_path, 'corn_prices.csv')
    group.to_csv(file_path, index=False)
```
